```python
import pandas as pd

base_path = "/content/drive/MyDrive/NASA Hackathon/Datasets/Final Datasets/"

# Initialize an empty dictionary to store results
nan_counts_by_year = {}

# Loop through each year from 2016 to 2023
for year in range(2016, 2024):
    file_path = f"{base_path}dsc_fc_summed_spectra_{year}_v01.csv"
    data = pd.read_csv(file_path)

    # Count the number of NaN values in each column
    nan_counts = data.isna().sum()

    nan_counts_by_year[year] = nan_counts.to_dict()

# Display results
for year, counts in nan_counts_by_year.items():
    print(f"NaN counts for {year}:\n")
    for column, count in counts.items():
        print(f"{column}: {count}")
    print("-" * 50)  # Print separator for clarity
```

# Data loading

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "/content/drive/MyDrive/NASA Hackathon/Datasets/Final Datasets/dsc_fc_summed_spectra_2020_v01.csv"
data = pd.read_csv(file_path)
data.head()
```

# Data final Steps

```python
def raw_flux_weighted_magnitude(data):
    """
    Compute the weighted magnitude of raw flux columns.

    Parameters:
    - data: DataFrame containing the raw flux columns.

    Returns:
    - Series of weighted magnitudes.
    """
    # Extract raw flux columns
    raw_flux_cols = [f"Raw_Flux_{i}" for i in range(1, 51)]

    # Compute the mean values for each raw flux column
    mean_values = data[raw_flux_cols].mean()

    # Compute the weighted magnitude
    weighted_magnitude = (data[raw_flux_cols] * mean_values).sum(axis=1)

    return weighted_magnitude

# Applying the function to the dataset
data['Raw_Flux_Weighted_Magnitude'] = raw_flux_weighted_magnitude(data)

# Display the first few rows to see the new column
print(data[['UTC_datetime', 'Raw_Flux_Weighted_Magnitude']].head())
```

```python
# Drop the raw flux columns
raw_flux_cols = [f"Raw_Flux_{i}" for i in range(1, 51)]
data_reduced = data.drop(columns=raw_flux_cols)
```

```python
data_reduced.head()
```

```python
import matplotlib.pyplot as plt

# Plotting the histogram
plt.figure(figsize=(10,6))
plt.hist(data_reduced['Raw_Flux_Weighted_Magnitude'], bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Raw_Flux_Weighted_Magnitude')
plt.xlabel('Weighted Magnitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# If the distribution of the Raw_Flux_Weighted_Magnitude column is Gaussian (or approximately normal), then the most appropriate scaling method is the StandardScaler. This scaler will transform the data such that it has a mean of 0 and a standard deviation of 1.

# Given a Gaussian distribution, the StandardScaler is a common choice because:

# It centers the data around zero, which is often beneficial for many machine learning algorithms.
# It maintains the shape of the original distribution (i.e., it won't distort the relative spacing between values).
# Algorithms that assume features to be on a similar scale and centered around zero (like linear regression, logistic regression, neural networks, and many others) will benefit from this scaling.
# Here's how you can apply StandardScaler to the Raw_Flux_Weighted_Magnitude column:


from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Reshape the data and fit the scaler
data_reduced['Raw_Flux_Weighted_Magnitude_Scaled'] = scaler.fit_transform(data_reduced['Raw_Flux_Weighted_Magnitude'].values.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt

# Plotting the histogram
plt.figure(figsize=(10,6))
plt.hist(data_reduced['Raw_Flux_Weighted_Magnitude_Scaled'], bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Raw_Flux_Weighted_Magnitude_Scaled')
plt.xlabel('Weighted Magnitude Scaled')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data_reduced = data_reduced.drop('Raw_Flux_Weighted_Magnitude',axis=1)
data_reduced.head()
```

```python
# Splitting the 'UTC_datetime' column into 'Date' and 'Time' columns
data_reduced[['Date', 'Time']] = data_reduced['UTC_datetime'].str.split(' ', expand=True)

# Dropping the original 'UTC_datetime' column
data_reduced = data_reduced.drop(columns=['UTC_datetime'])

# Display the first few rows to see the changes
data_reduced.head()
```

```python
# Convert the 'Time' column to datetime format
data_reduced['Time'] = pd.to_datetime(data_reduced['Time'], format='%H:%M:%S').dt.time

# Extract the hour from the 'Time' column
data_reduced['Hour'] = data_reduced['Time'].apply(lambda x: x.hour)

# Create the 'group' column based on the hour
data_reduced['group'] = (data_reduced['Hour'] // 3) + 1

# Drop the temporary 'Hour' column
data_reduced = data_reduced.drop(columns=['Hour'])

# Display the first few rows to see the changes
data_reduced.head()
```

```python
# the desired column order
desired_order = ['Date', 'Time', 'group', 'Bx_GSE', 'By_GSE', 'Bz_GSE', 'Raw_Flux_Weighted_Magnitude_Scaled']

# Rearrange the columns
data_reduced = data_reduced[desired_order]

# Display the first few rows to see the rearranged columns
data_reduced
```
