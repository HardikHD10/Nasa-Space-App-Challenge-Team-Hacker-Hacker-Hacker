```python
import pandas as pd

# Define column names
column_names = ["UTC_datetime"]

# Magnetic field vector components in GSE
column_names += ["Bx_GSE", "By_GSE", "Bz_GSE"]

# Raw measurement spectrum from Faraday cup
for i in range(1, 51):
    column_names.append(f"Raw_Flux_{i}")

# Load the dataset with the new column names
file_path = "/content/drive/MyDrive/NASA Hackathon/Datasets/Intial Datasets/dsc_fc_summed_spectra_2016_v01.csv"
dscovr_data = pd.read_csv(file_path, header=None, names=column_names)

# Display the first few rows of the dataset
print(dscovr_data.head())
```

```python
# Define the path to save the processed dataset
save_path = "/content/drive/MyDrive/NASA Hackathon/Datasets/Final Datasets/dsc_fc_summed_spectra_2016_v01.csv"

# Save the dataset to the new CSV file
dscovr_data.to_csv(save_path, index=False)

print(f"Dataset saved to {save_path}")
```
