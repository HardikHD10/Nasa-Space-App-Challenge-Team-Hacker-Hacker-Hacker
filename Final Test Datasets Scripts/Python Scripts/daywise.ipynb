```python
# Reading the dataset using the provided script with delim_whitespace=True
data4 = pd.read_csv("/content/drive/MyDrive/NASA Hackathon/Final Test Datasets Scripts/Raw Dataset/www-app3.gfz-potsdam.de_kp_index_Kp_ap_Ap_SN_F107_since_1932.txt", delim_whitespace=True)

# Extracting the column names
data4.columns
```

```python
# Reading the provided dataset using the given script
data1 = pd.read_csv("/content/drive/MyDrive/NASA Hackathon/Final Test Datasets Scripts/Raw Dataset/rtsw_plot_data_2023-08-17T00_00_00 (1).txt", delim_whitespace=True, header=0, index_col=None)

# Combining the index (which has the date) and the "Timestamp" column (which has the time) into a single column
data1["Timestamp"] = data1.index.astype(str) + " " + data1["Timestamp"]

# Resetting the index to default integers
data1 = data1.reset_index(drop=True)

# Extracting the column names for the provided dataset
data1.columns
```

```python
data1
```

```python
# Step 1: Load the datasets
# We have already loaded the common dataset previously as data4

# Extracting the Timestamp column from data1
timestamp_column = data1["Timestamp"]

# Displaying the first few rows of the Timestamp column
timestamp_column.head()
```

```python
# Correcting the Timestamp column to only have the datetime values
data1["Timestamp"] = pd.to_datetime(data1["Timestamp"].str.split().str[-2] + " " + data1["Timestamp"].str.split().str[-1])

# Extracting date and time separately in desired formats
data1['Date'] = data1['Timestamp'].dt.strftime('%d-%m-%Y')
data1['Time'] = data1['Timestamp'].dt.strftime('%H:%M:%S')

# Displaying the first few rows of the corrected Date and Time columns
data1[['Date', 'Time']].head()
```

```python
data1
```

```python
# Ensure the "Date" column in both dataframes is of datetime format
data4["Date"] = pd.to_datetime(data4["YYY"].astype(str) + "-" + data4["MM"].astype(str).str.zfill(2) + "-" + data4["DD"].astype(str).str.zfill(2))
data1["Date"] = pd.to_datetime(data1["Date"])

# Merge the specified columns from data4 into data1 based on the "Date" column
merged_data = pd.merge(data1, data4[['Date', 'Ap', 'SN', 'F10.7obs', 'F10.7adj']], on='Date', how='left')
```

```python
merged_data
```

```python
merged_data['ap_value'] = merged_data['Ap']
```

```python
merged_data
```

```python
data = merged_data.drop(["Ap"],axis=1)
```

```python
# Replace NaN values with group median
for column in data.columns:
    if data[column].dtype != 'object':
        data[column].fillna(data[column].median(), inplace=True)
```

```python
data
```

```python
import numpy as np
```

```python
data.columns
```

```python
# List of numerical columns based on the provided columns
numerical_columns = ['Bt-med', 'Bt-min', 'Bt-max', 'Bx-med', 'Bx-min',
                     'Bx-max', 'By-med', 'By-min', 'By-max', 'Bz-med', 'Bz-min', 'Bz-max',
                     'Phi-mean', 'Phi-min', 'Phi-max', 'Theta-med', 'Theta-min', 'Theta-max',
                     'Dens-med', 'Dens-min', 'Dens-max', 'Speed-med', 'Speed-min',
                     'Speed-max', 'Temp-med', 'Temp-min', 'Temp-max', 'SN',
                     'F10.7obs', 'F10.7adj', 'ap_value']

# Replace outliers based on the 25th and 75th quantiles
for column in numerical_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    data[column] = np.where(data[column] < Q1, Q1, data[column])
    data[column] = np.where(data[column] > Q3, Q3, data[column])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler

# Standard scaling on the numerical columns
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
data
```

```python
# Extracting the specified columns to finalize the X_data
X_data = data[['Date', 'Time','Bt-min', 'Bx-med', 'By-med', 'Phi-mean', 'Phi-min', 'Phi-max',
       'Theta-med', 'Theta-min', 'Theta-max', 'Dens-max', 'Speed-min',
       'Temp-med', 'Temp-min', 'Temp-max', 'SN', 'ap_value']]

# Displaying the first few rows of the finalized X_data
X_data.head()
```

```python
# Verifying the presence and non-null counts of the specified columns in X_data
column_verification = X_data[['Bt-min', 'Bx-med', 'By-med', 'Phi-mean', 'Phi-min', 'Phi-max',
       'Theta-med', 'Theta-min', 'Theta-max', 'Dens-max', 'Speed-min',
       'Temp-med', 'Temp-min', 'Temp-max', 'SN', 'ap_value']].info()

column_verification
```

```python
# Reordering the columns in X_data according to the specified sequence
ordered_columns = ['Bt-min', 'Bx-med', 'By-med', 'Phi-mean', 'Phi-min', 'Phi-max',
       'Theta-med', 'Theta-min', 'Theta-max', 'Dens-max', 'Speed-min',
       'Temp-med', 'Temp-min', 'Temp-max', 'SN', 'ap_value']

X_data = X_data[['Date', 'Time'] + ordered_columns]

# Displaying the first few rows of the reordered X_data
X_data.head()
```
